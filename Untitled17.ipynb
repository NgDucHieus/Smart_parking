1. Hàm cho Webcam (sử dụng Haar Cascade)

```python
def js_to_image_cv(js_reply_img):
    try:
        image_data = js_reply_img.split(',')[1]
        image_bytes = base64.b64decode(image_data)
        jpg_as_np = np.frombuffer(image_bytes, dtype=np.uint8)
        img = cv2.imdecode(jpg_as_np, flags=cv2.IMREAD_COLOR) # Trả về ảnh BGR của OpenCV
        return img
    except Exception as e:
        print(f"Lỗi js_to_image_cv: {e}")
        return None

def bbox_to_bytes_cv(bbox_array): # bbox_array là ảnh RGBA đã vẽ bounding box
    try:
        img_pil = Image.fromarray(bbox_array, 'RGBA')
        buffer = io.BytesIO()
        img_pil.save(buffer, format='PNG')
        byte_data = buffer.getvalue()
        base64_str = base64.b64encode(byte_data).decode('utf-8')
        return 'data:image/png;base64,' + base64_str
    except Exception as e:
        print(f"Lỗi bbox_to_bytes_cv: {e}")
        return ""
```

2. Hàm cho FaceNet (sử dụng MTCNN)

```python
def extract_face_for_facenet(image_path_or_pil_image, required_size=(160, 160), confidence_threshold=0.95):
    face_arrays = []
    try:
        if isinstance(image_path_or_pil_image, str): # Nếu là đường dẫn file
            img = Image.open(image_path_or_pil_image)
        elif isinstance(image_path_or_pil_image, Image.Image): # Nếu là đối tượng PIL Image
             img = image_path_or_pil_image
        else:
            print("Đầu vào không hợp lệ cho extract_face_for_facenet.")
            return []

        if img.mode != 'RGB':
            img = img.convert('RGB')
        pixels = np.asarray(img)
    except FileNotFoundError:
        print(f"Lỗi: Không tìm thấy tệp ảnh tại '{image_path_or_pil_image}'")
        return face_arrays
    except Exception as e:
        print(f"Lỗi khi đọc hoặc xử lý ảnh {image_path_or_pil_image}: {e}")
        return face_arrays

    results = mtcnn_detector.detect_faces(pixels)
    for result in results:
        if result['confidence'] < confidence_threshold:
            continue
        x1, y1, width, height = result['box']
        x1, y1 = abs(x1), abs(y1)
        x2, y2 = x1 + width, y1 + height
        x1,y1,x2,y2 = max(0,x1), max(0,y1), min(pixels.shape[1],x2), min(pixels.shape[0],y2)
        if x2 <= x1 or y2 <= y1: continue
        face = pixels[y1:y2, x1:x2]
        try:
            image_pil = Image.fromarray(face)
            image_pil = image_pil.resize(required_size)
            face_arrays.append(np.asarray(image_pil))
        except Exception as e_resize:
            print(f"Lỗi resize (FaceNet): {e_resize}")
            continue
    if not face_arrays and isinstance(image_path_or_pil_image, str):
        print(f"Không tìm thấy khuôn mặt nào (MTCNN) trong {image_path_or_pil_image}")
    return face_arrays

def get_facenet_embeddings(list_of_face_arrays):
    if not list_of_face_arrays: return []
    try:
        faces_pixels = np.asarray(list_of_face_arrays, 'float32')
        embeddings = facenet_embedder.embeddings(faces_pixels)
        return embeddings
    except Exception as e:
        print(f"Lỗi khi trích xuất embeddings FaceNet: {e}")
        return []

print("\nCác hàm tiện ích đã được định nghĩa.")
```

```python
dataset_base_path = '/content/drive/MyDrive/BTL cuối kỳ AI/FACE ID/Dataset/'
os.makedirs(dataset_base_path, exist_ok=True)

def capture_faces_via_webcam(person_name, base_dataset_path, num_images_to_save=20):
    """
    Bật webcam, phát hiện khuôn mặt bằng Haar Cascade và lưu ảnh.
    """
    person_save_folder = os.path.join(base_dataset_path, person_name)
    os.makedirs(person_save_folder, exist_ok=True)
    print(f"Sẽ lưu ảnh cho '{person_name}' vào: {person_save_folder}")

    # JavaScript cho webcam
    js = Javascript('''
        var video; var div = null; var stream; var captureCanvas; var imgElement;
        var labelElement; var pendingResolve = null; var shutdown = false;

        function removeDom() {
          if (stream) { stream.getVideoTracks().forEach(track => track.stop()); }
          if (video) video.remove(); if (div) div.remove();
          video = null; div = null; stream = null; imgElement = null; captureCanvas = null; labelElement = null;
        }

        function onAnimationFrame() {
          if (!shutdown) { window.requestAnimationFrame(onAnimationFrame); }
          if (pendingResolve) {
            var result = "";
            if (!shutdown && video && video.readyState === 4 && video.videoWidth > 0) {
              captureCanvas.width = video.videoWidth; captureCanvas.height = video.videoHeight;
              captureCanvas.getContext('2d').drawImage(video, 0, 0, video.videoWidth, video.videoHeight);
              result = captureCanvas.toDataURL('image/jpeg', 0.8);
            }
            var lp = pendingResolve; pendingResolve = null; lp(result);
          }
        }

        async function createDom() {
          if (div !== null) { return stream; }
          shutdown = false; div = document.createElement('div');
          div.style.border = '2px solid black'; div.style.padding = '3px';
          div.style.width = '100%'; div.style.maxWidth = '650px'; document.body.appendChild(div);

          const modelOut = document.createElement('div');
          modelOut.innerHTML = "<span>Trạng thái Webcam: </span>";
          labelElement = document.createElement('span'); labelElement.innerText = 'Đang khởi tạo...';
          labelElement.style.fontWeight = 'bold'; modelOut.appendChild(labelElement); div.appendChild(modelOut);

          video = document.createElement('video'); video.style.display = 'block';
          video.setAttribute('playsinline', ''); video.onclick = () => { shutdown = true; };
          stream = await navigator.mediaDevices.getUserMedia({video: {facingMode: "user", width: {ideal:640}, height: {ideal:480}}});
          div.appendChild(video);

          imgElement = document.createElement('img'); imgElement.style.position = 'absolute';
          imgElement.style.zIndex = 1; div.appendChild(imgElement);

          const instruction = document.createElement('div');
          instruction.innerHTML = '<span style="color: red; font-weight: bold;">NHẤP VIDEO ĐỂ DỪNG SỚM.</span>';
          instruction.style.cursor = 'pointer'; div.appendChild(instruction);
          instruction.onclick = () => { shutdown = true; };

          video.srcObject = stream; await video.play();
          captureCanvas = document.createElement('canvas');
          window.requestAnimationFrame(onAnimationFrame);
          google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);
          if (labelElement) labelElement.innerText = 'Camera sẵn sàng!';
          return stream;
        }

        async function stream_frame_webcam(label, imgData) {
          if (shutdown) { removeDom(); return ''; }
          await createDom();
          if (labelElement && label != "") { labelElement.innerHTML = label; }
          if (imgElement && video) {
            var videoRect = video.getBoundingClientRect();
            imgElement.style.top = videoRect.top + "px"; imgElement.style.left = videoRect.left + "px";
            imgElement.style.width = videoRect.width + "px"; imgElement.style.height = videoRect.height + "px";
            if (imgData != "") { imgElement.src = imgData; }
            else { imgElement.src = "data:image/gif;base64,R0lGODlhAQABAIAAAAAAAP///yH5BAEAAAAALAAAAAABAAEAAAIBRAA7"; }
          }
          var result = await new Promise(function(resolve) { pendingResolve = resolve; });
          if (!result) { removeDom(); return ''; }
          return {'img': result};
        }
    ''')
    display(js)

    def video_frame_webcam(label, bbox_data_url):
        data = eval_js('stream_frame_webcam("{}", "{}")'.format(label, bbox_data_url))
        return data

    print(f"\nBắt đầu thu thập ảnh cho '{person_name}' qua webcam...")
    print("Trình duyệt sẽ yêu cầu quyền truy cập camera. Vui lòng đồng ý.")
    # video_stream_webcam() # JS được display ở trên rồi

    label_html = 'Đang tìm khuôn mặt...'
    bbox_data_url = ''
    saved_count_for_person = 0
    frame_count = 0
    max_capture_frames = 600 # Tối đa số frame xử lý cho việc chụp

    try:
        while saved_count_for_person < num_images_to_save and frame_count < max_capture_frames:
            js_reply = video_frame_webcam(label_html, bbox_data_url)
            if not js_reply or not js_reply.get("img"):
                print("Luồng webcam đã dừng hoặc không nhận được ảnh.")
                break

            img_bgr_cv = js_to_image_cv(js_reply["img"])
            if img_bgr_cv is None:
                frame_count += 1
                bbox_data_url = "" # Reset overlay
                time.sleep(0.1)
                continue

            overlay_height_cv, overlay_width_cv, _ = img_bgr_cv.shape
            bbox_overlay_rgba_cv = np.zeros((overlay_height_cv, overlay_width_cv, 4), dtype=np.uint8)
            gray_cv = cv2.cvtColor(img_bgr_cv, cv2.COLOR_BGR2GRAY)
            faces_cv = face_cascade_cv.detectMultiScale(gray_cv, 1.15, 5, minSize=(80,80))

            detected_this_frame = False
            if len(faces_cv) > 0:
                detected_this_frame = True
                (x, y, w, h) = faces_cv[0] # Lấy khuôn mặt lớn nhất hoặc đầu tiên
                cv2.rectangle(bbox_overlay_rgba_cv, (x, y), (x+w, y+h), (255,0,0,255), 2) # Blue

                face_roi_cv = img_bgr_cv[y:y+h, x:x+w]
                if face_roi_cv.size > 0:
                    timestamp = int(time.time() * 1000)
                    filename = os.path.join(person_save_folder, f"{person_name.replace(' ','_')}_{timestamp}.jpg")
                    cv2.imwrite(filename, face_roi_cv)
                    saved_count_for_person += 1
                    print(f"Đã lưu ảnh {saved_count_for_person}/{num_images_to_save} cho {person_name}: {filename}")
                    # Cho người dùng chút thời gian để thay đổi vị trí nếu muốn
                    if saved_count_for_person < num_images_to_save: time.sleep(0.5)


            alpha_cv = bbox_overlay_rgba_cv[:,:,3]
            alpha_cv[np.any(bbox_overlay_rgba_cv[:,:,:3] != [0,0,0], axis=2)] = 255
            bbox_overlay_rgba_cv[:,:,3] = alpha_cv
            bbox_data_url = bbox_to_bytes_cv(bbox_overlay_rgba_cv)

            label_html = f"Cho {person_name}: Đã lưu {saved_count_for_person}/{num_images_to_save}. Tìm thấy: {len(faces_cv)>0}"
            frame_count += 1
            if saved_count_for_person >= num_images_to_save:
                print(f"Đã thu thập đủ {num_images_to_save} ảnh cho {person_name}.")
                break

    except Exception as e:
        print(f"Lỗi trong quá trình thu thập ảnh webcam: {type(e).__name__}: {e}")
    finally:
        try:
            eval_js('removeDom()')
            print("Đã đóng luồng webcam.")
        except:
            pass
    print(f"Kết thúc thu thập ảnh cho {person_name}. Đã lưu {saved_count_for_person} ảnh.")
```

VI. NHẬN DẠNG KHUÔN MẶT BẰNG FACENET

```python
import os
import cv2
import numpy as np
import time
from IPython.display import display, Javascript
from google.colab.output import eval_js
from PIL import Image
import io
import base64
import pickle
from sklearn.metrics.pairwise import cosine_similarity


# --- PHẦN 1: CÀI ĐẶT VÀ TẢI TÀI NGUYÊN (NGƯỜI DÙNG CẦN THAY ĐỔI) ---

# 1.1. Khởi tạo MTCNN detector
try:
    from mtcnn.mtcnn import MTCNN
    detector_mtcnn = MTCNN()
    print("MTCNN detector đã được khởi tạo thành công.")
except ImportError:
    print("Lỗi: Thư viện MTCNN chưa được cài đặt. Vui lòng cài đặt: pip install mtcnn")
    detector_mtcnn = None
except Exception as e:
    print(f"Lỗi khi khởi tạo MTCNN: {e}")
    detector_mtcnn = None

# 1.2. Tải mô hình FaceNet của bạn
#      QUAN TRỌNG: BẠN PHẢI CUNG CẤP CODE ĐỂ TẢI MÔ HÌNH FACENET THỰC TẾ CỦA MÌNH Ở ĐÂY.
#      Ví dụ:
from keras_facenet import FaceNet
facenet_model = FaceNet() # Thay thế bằng cách tải model của bạn
print("Mô hình Keras-FaceNet đã được tải.")
#
#      Hoặc nếu bạn có tệp .h5:
#      from tensorflow.keras.models import load_model
#      try:
#          facenet_model = load_model('YOUR_PATH_TO_FACENET_MODEL.h5') # THAY ĐƯỜNG DẪN
#          print("Mô hình FaceNet tùy chỉnh đã được tải từ tệp.")
#      except Exception as e:
#          print(f"Lỗi khi tải mô hình FaceNet tùy chỉnh: {e}")
#          facenet_model = None

# 1.3. Hàm trích xuất embedding FaceNet
def get_facenet_embeddings(face_pixels_list_rgb, model):
    if model is None:
        print("Lỗi trong get_facenet_embeddings: Mô hình FaceNet chưa được cung cấp (model is None).")
        return [] # Trả về list rỗng
    if not face_pixels_list_rgb:
        return []


    try:
        samples = np.asarray(face_pixels_list_rgb, 'float32')
        # QUAN TRỌNG: Tiền xử lý pixel (chuẩn hóa) PHẢI khớp với mô hình FaceNet của bạn.
        # Nhiều thư viện (như Keras-FaceNet) có thể tự xử lý.
        # Ví dụ (nếu cần chuẩn hóa thủ công):
        # samples = (samples - 127.5) / 127.5 # Về khoảng [-1, 1]


        # --- THAY THẾ LOGIC NÀY BẰNG CÁCH GỌI MÔ HÌNH FACENET THỰC TẾ CỦA BẠN ---
        if hasattr(model, 'embeddings'): # Cho Keras-FaceNet
            embeddings = model.embeddings(samples)
            return embeddings
        elif hasattr(model, 'predict'): # Cho mô hình Keras chuẩn
            embeddings = model.predict(samples)
            return embeddings
        else:
            print("CẢNH BÁO get_facenet_embeddings: Không biết cách dùng `model`. Trả về embeddings GIẢ LẬP.")
            return [np.random.rand(1, (model.output_shape[-1] if hasattr(model, "output_shape") and model.output_shape[-1] else 512)).astype('float32') for _ in face_pixels_list_rgb]

    except Exception as e:
        print(f"Lỗi trong quá trình trích xuất embedding: {e}")
        import traceback
        traceback.print_exc()
        return [] # Trả về list rỗng




# --- Hàm tiện ích được sử dụng trong Phần 1.4 (định nghĩa extract_faces_details_mtcnn ở Phần 3) ---
def extract_face_from_image_path(image_path, detector, required_size=(160, 160)):
    img_bgr = cv2.imread(image_path)
    if img_bgr is None:
        return []
    face_details = extract_faces_details_mtcnn(img_bgr, detector, required_size) # extract_faces_details_mtcnn định nghĩa ở Phần 3
    return [detail['face_image_rgb_resized'] for detail in face_details]




# 1.4. Tải hoặc Xây dựng (và Lưu) Cơ sở dữ liệu Khuôn mặt
dataset_base_path = '/content/drive/MyDrive/BTL cuối kỳ AI/FACE ID/Dataset/' # KIỂM TRA LẠI ĐƯỜNG DẪN NÀY
database_file_path = "face_embeddings_database.pkl"
face_database_facenet = {}


if os.path.exists(database_file_path):
    try:
        with open(database_file_path, 'rb') as f:
            face_database_facenet = pickle.load(f)
        print(f"Đã tải thành công cơ sở dữ liệu FaceNet từ: {database_file_path}")
        if face_database_facenet:
            print(f"Thông tin database đã tải: { {name: len(embs) for name, embs in face_database_facenet.items()} }")
        else: # File tồn tại nhưng rỗng hoặc không đúng định dạng
            print(f"Tệp database '{database_file_path}' rỗng hoặc không hợp lệ. Sẽ thử xây dựng lại.")
            face_database_facenet = {} # Reset để xây dựng lại
    except Exception as e:
        print(f"Lỗi khi tải cơ sở dữ liệu từ '{database_file_path}': {e}. Sẽ thử xây dựng lại.")
        face_database_facenet = {}


if not face_database_facenet: # Chỉ xây dựng nếu chưa tải được hoặc database rỗng
    print(f"\nBắt đầu xây dựng cơ sở dữ liệu embeddings FaceNet từ: {dataset_base_path}")
    if not os.path.exists(dataset_base_path):
        print(f"LỖI: Đường dẫn dataset '{dataset_base_path}' không tồn tại.")
    elif detector_mtcnn is None or facenet_model is None: # Cần cả hai model để xây dựng
        print("LỖI: MTCNN detector hoặc FaceNet model (trong `facenet_model`) chưa được khởi tạo. Không thể xây dựng database.")
    else:
        for person_name in os.listdir(dataset_base_path):
            person_folder_path = os.path.join(dataset_base_path, person_name)
            if os.path.isdir(person_folder_path):
                print(f"  Đang xử lý cho (FaceNet DB): {person_name}")
                person_embeddings_list = []
                for image_file in os.listdir(person_folder_path):
                    image_path_full = os.path.join(person_folder_path, image_file)
                    if image_file.lower().endswith(('.png', '.jpg', '.jpeg')):
                        face_arrays_rgb = extract_face_from_image_path(image_path_full, detector_mtcnn)
                        if face_arrays_rgb:
                            embeddings_list = get_facenet_embeddings(face_arrays_rgb, facenet_model)
                            if embeddings_list: # Kiểm tra embeddings_list không rỗng
                                person_embeddings_list.extend(embeddings_list)

                if person_embeddings_list:
                    face_database_facenet[person_name] = person_embeddings_list
                    print(f"  -> FaceNet DB: Đã tạo {len(person_embeddings_list)} embedding(s) cho {person_name}.")
                else:
                    print(f"  -> FaceNet DB: Không tìm thấy embedding nào hợp lệ cho {person_name} từ các ảnh.")

        print("\nHoàn tất xây dựng cơ sở dữ liệu embeddings FaceNet từ dataset.")
        if face_database_facenet:
            try:
                with open(database_file_path, 'wb') as f:
                    pickle.dump(face_database_facenet, f)
                print(f"Đã lưu cơ sở dữ liệu FaceNet mới xây dựng vào: {database_file_path}")
            except Exception as e:
                print(f"Lỗi khi lưu cơ sở dữ liệu FaceNet mới xây dựng: {e}")
        else:
            print("Cơ sở dữ liệu xây dựng từ dataset rỗng, không có gì để lưu.")


if not face_database_facenet:
    print("CẢNH BÁO CUỐI CÙNG: Cơ sở dữ liệu `face_database_facenet` vẫn trống.")
else:
    print("\nCơ sở dữ liệu FaceNet `face_database_facenet` đã sẵn sàng:")
    for name, embs in face_database_facenet.items():
        print(f"- {name}: {len(embs)} embedding(s)")




# --- PHẦN 2: CÁC HÀM TIỆN ÍCH ---
def js_to_image_cv(js_reply_img):
    if not js_reply_img: return None
    try:
        header, encoded = js_reply_img.split(',', 1)
        jpeg_bytes = io.BytesIO(base64.b64decode(encoded))
        pil_image = Image.open(jpeg_bytes)
        return cv2.cvtColor(np.array(pil_image), cv2.COLOR_RGB2BGR)
    except Exception as e:
        # print(f"Lỗi chuyển đổi js_to_image_cv: {e}") # Giảm log
        return None


def bbox_to_bytes_cv(bbox_overlay_rgba_cv):
    if bbox_overlay_rgba_cv is None: return ""
    try:
        _, buffer = cv2.imencode('.png', bbox_overlay_rgba_cv)
        png_as_base64 = base64.b64encode(buffer).decode('ascii')
        return 'data:image/png;base64,{}'.format(png_as_base64)
    except Exception as e:
        # print(f"Lỗi chuyển đổi bbox_to_bytes_cv: {e}") # Giảm log
        return ""


# --- PHẦN 3: HÀM TRÍCH XUẤT KHUÔN MẶT BẰNG MTCNN (CHO LIVE WEBCAM VÀ DATABASE BUILDING) ---
def extract_faces_details_mtcnn(bgr_frame, detector, required_size=(160, 160)):
    if bgr_frame is None: return []
    if detector is None:
        print("Lỗi trong extract_faces_details_mtcnn: MTCNN detector chưa được cung cấp.")
        return []
    rgb_frame = cv2.cvtColor(bgr_frame, cv2.COLOR_BGR2RGB)
    try:
        detections = detector.detect_faces(rgb_frame)
    except Exception as e:
        # print(f"Lỗi khi gọi MTCNN detect_faces: {e}") # Giảm log
        return []
    extracted_faces_details = []
    for det in detections:
        if det['confidence'] < 0.90: continue
        x, y, w, h = det['box']
        if w <= 0 or h <= 0: continue
        x, y = abs(x), abs(y)
        face_rgb = rgb_frame[y:y+h, x:x+w]
        if face_rgb.size == 0: continue
        try:
            face_image_resized_rgb = cv2.resize(face_rgb, required_size, interpolation=cv2.INTER_AREA)
            extracted_faces_details.append({
                'box': (x, y, w, h),
                'confidence': det['confidence'],
                'face_image_rgb_resized': face_image_resized_rgb
            })
        except Exception: # Bỏ qua khuôn mặt này nếu resize lỗi
            continue
    return extracted_faces_details


# --- PHẦN 4: HÀM CHÍNH CHO NHẬN DẠNG TRỰC TIẾP ---
def recognize_faces_live_facenet(facenet_db,
                                 threshold=0.55,
                                 mtcnn_detector_obj=None,
                                 facenet_embedding_model_obj=None):
    if mtcnn_detector_obj is None:
        print("LỖI CHẠY: MTCNN detector chưa được cung cấp.")
        return
    if facenet_embedding_model_obj is None: # Model này được dùng bởi get_facenet_embeddings
        print("CẢNH BÁO CHẠY: `facenet_embedding_model_obj` (Mô hình FaceNet) chưa được cung cấp.")
        print("         Nếu `get_facenet_embeddings` không có model, nó sẽ trả về dữ liệu giả lập.")
    if not facenet_db:
        print("LỖI CHẠY: Cơ sở dữ liệu FaceNet trống.")
        return


    print("\nKhởi tạo webcam cho nhận dạng FaceNet trực tiếp...")
    js_code = Javascript('''
        var video; var div = null; var stream; var captureCanvas; var imgElement;
        var labelElement; var pendingResolve = null; var shutdown = false;


        function removeDom() {
          if (stream) { stream.getVideoTracks().forEach(track => track.stop()); }
          if (video) video.remove(); if (div) div.remove();
          video = null; div = null; stream = null; imgElement = null; captureCanvas = null; labelElement = null;
          console.log("JS: DOM elements removed and stream stopped.");
        }
        function userShutdown() { shutdown = true; console.log("JS: Shutdown initiated by user click."); removeDom(); /* Gọi removeDom ở đây */}


        function onAnimationFrame() {
          if (!shutdown) { window.requestAnimationFrame(onAnimationFrame); }
          if (pendingResolve) {
            var result = "";
            if (!shutdown && video && video.readyState === 4 && video.videoWidth > 0) {
              captureCanvas.width = video.videoWidth; captureCanvas.height = video.videoHeight;
              captureCanvas.getContext('2d').drawImage(video, 0, 0, video.videoWidth, video.videoHeight);
              result = captureCanvas.toDataURL('image/jpeg', 0.8);
            }
            var lp = pendingResolve; pendingResolve = null; lp(result);
          }
        }
        async function createDom() {
          if (div !== null) { return stream; }
          shutdown = false; div = document.createElement('div');
          div.style.border = '2px solid black'; div.style.padding = '3px';
          div.style.width = '100%'; div.style.maxWidth = '650px'; document.body.appendChild(div);


          const modelOut = document.createElement('div');
          modelOut.innerHTML = "<span>Trạng thái Webcam: </span>";
          labelElement = document.createElement('span'); labelElement.innerText = 'Đang khởi tạo...';
          labelElement.style.fontWeight = 'bold'; modelOut.appendChild(labelElement); div.appendChild(modelOut);


          video = document.createElement('video'); video.style.display = 'block';
          video.setAttribute('playsinline', ''); video.onclick = userShutdown;
          try {
            stream = await navigator.mediaDevices.getUserMedia({video: {facingMode: "user", width: {ideal:640}, height: {ideal:480}}});
          } catch (e) {
            console.error("JS: Lỗi khi getUserMedia: ", e);
            if (labelElement) labelElement.innerText = 'Lỗi camera: ' + e.message;
            removeDom(); // Dọn dẹp nếu không lấy được stream
            return null; // Trả về null nếu không có stream
          }
          div.appendChild(video);


          imgElement = document.createElement('img'); imgElement.style.position = 'absolute';
          imgElement.style.zIndex = 1; div.appendChild(imgElement);


          const instruction = document.createElement('div');
          instruction.innerHTML = '<span style="color: red; font-weight: bold;">NHẤP VIDEO ĐỂ DỪNG.</span>';
          instruction.style.cursor = 'pointer'; instruction.onclick = userShutdown; div.appendChild(instruction);


          video.srcObject = stream;
          try {
            await video.play();
          } catch (e_play) {
            console.error("JS: Lỗi khi play video: ", e_play);
            if (labelElement) labelElement.innerText = 'Lỗi play video: ' + e_play.message;
            removeDom();
            return null;
          }
          captureCanvas = document.createElement('canvas');
          window.requestAnimationFrame(onAnimationFrame);
          if (google && google.colab && google.colab.output && google.colab.output.setIframeHeight) {
            google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);
          }
          if (labelElement) labelElement.innerText = 'Camera sẵn sàng!';
          console.log("JS: DOM created and camera ready.");
          return stream;
        }
        async function stream_frame_webcam(label, imgData) {
          if (shutdown) { /*removeDom(); không cần gọi ở đây nữa vì userShutdown đã gọi*/ return ''; }
          const currentStream = await createDom();
          if (!currentStream) { // Nếu createDom trả về null (lỗi camera)
             shutdown = true; // Đảm bảo Python loop cũng dừng
             removeDom();
             return '';
          }
          if (labelElement && label != "") { labelElement.innerHTML = label; }
          if (imgElement && video && video.offsetParent) {
            var videoRect = video.getBoundingClientRect();
            imgElement.style.top = videoRect.top + "px"; imgElement.style.left = videoRect.left + "px";
            imgElement.style.width = videoRect.width + "px"; imgElement.style.height = videoRect.height + "px";
            if (imgData && imgData !== "") { imgElement.src = imgData; }
            else { imgElement.src = "data:image/gif;base64,R0lGODlhAQABAIAAAAAAAP///yH5BAEAAAAALAAAAAABAAEAAAIBRAA7"; }
          }
          var result = await new Promise(function(resolve) { pendingResolve = resolve; });
          if (shutdown) { /*removeDom();*/ return ''; }
          return {'img': result};
        }
    ''')
    display(js_code)


    def video_frame_webcam_py(label_str, bbox_data_url_str):
        js_command = 'stream_frame_webcam("{}", "{}")'.format(label_str.replace('"', '\\"'), bbox_data_url_str)
        try:
            data = eval_js(js_command)
            # Nếu JS trả về chuỗi rỗng, nghĩa là đã shutdown
            if isinstance(data, str) and data == '':
                return None # Python hiểu là đã shutdown
            return data
        except Exception:
            return None # Lỗi hoặc DOM đã bị xóa


    print("Trình duyệt sẽ yêu cầu quyền truy cập camera. Vui lòng đồng ý.")
    print("Nhấn vào video trong Colab output để dừng chương trình.")
    current_label_html = 'Đang tìm khuôn mặt...'
    current_bbox_data_url = ''


    try:
        while True:
            js_reply_data = video_frame_webcam_py(current_label_html, current_bbox_data_url)
            if not js_reply_data:
                print("Luồng webcam đã dừng (JS không trả về dữ liệu hoặc đã shutdown).")
                break
            img_bgr_frame = js_to_image_cv(js_reply_data.get("img"))


            if img_bgr_frame is None: # Có thể do js_reply_data['img'] rỗng hoặc lỗi chuyển đổi
                # Kiểm tra xem JS có còn hoạt động không
                try:
                    if eval_js('shutdown'): # Nếu biến shutdown trong JS là true
                        print("Luồng webcam đã dừng do người dùng click (JS shutdown=true).")
                        break
                except Exception: # Lỗi nếu DOM đã bị xóa
                     print("Không thể kiểm tra biến JS 'shutdown' (DOM có thể đã bị xóa). Luồng dừng.")
                     break
                # print("Không nhận được khung hình hợp lệ, thử lại...") # Giảm log
                time.sleep(0.05)
                current_bbox_data_url = ""
                continue


            overlay_h, overlay_w, _ = img_bgr_frame.shape
            bbox_overlay_img_rgba = np.zeros((overlay_h, overlay_w, 4), dtype=np.uint8)
            detected_faces_info = extract_faces_details_mtcnn(img_bgr_frame, mtcnn_detector_obj)
            frame_recognition_labels = []


            if not detected_faces_info:
                current_label_html = "Đang tìm (MTCNN)..."
            else:
                face_arrays_to_embed = [info['face_image_rgb_resized'] for info in detected_faces_info]
                if face_arrays_to_embed:
                    unknown_face_embeddings = get_facenet_embeddings(face_arrays_to_embed, facenet_embedding_model_obj)


                    # --- ĐOẠN CODE ĐÃ SỬA LỖI VALUEERROR ---
                    if unknown_face_embeddings is not None and \
                       hasattr(unknown_face_embeddings, '__len__') and \
                       len(unknown_face_embeddings) > 0: # Kiểm tra có embeddings trả về


                        if len(unknown_face_embeddings) != len(detected_faces_info):
                            # print(f"Cảnh báo: Số embeddings ({len(unknown_face_embeddings)}) không khớp số khuôn mặt ({len(detected_faces_info)}).")
                            pass # Sẽ không có lỗi nếu chỉ một vài khuôn mặt có embedding


                        for i_emb, emb_data in enumerate(unknown_face_embeddings):
                            # Tìm thông tin khuôn mặt tương ứng (nếu số lượng khớp, i_emb sẽ là index)
                            # Nếu số lượng không khớp, logic này cần cẩn thận hơn, nhưng hiện tại giả định get_embeddings trả về đủ hoặc không trả về gì cho 1 face array.
                            if i_emb >= len(detected_faces_info): continue # An toàn nếu get_embeddings trả về nhiều hơn dự kiến


                            current_emb = emb_data # Giả sử mỗi phần tử là một embedding
                            if current_emb is None or (hasattr(current_emb, 'size') and current_emb.size == 0):
                                # print(f"Cảnh báo: Embedding thứ {i_emb} rỗng hoặc None.")
                                continue

                            current_face_info = detected_faces_info[i_emb]
                            (fx, fy, fw, fh) = current_face_info['box']
                            best_match_identity = "Unknown"
                            max_similarity_score = 0.0

                            try:
                                current_emb_reshaped = np.asarray(current_emb).reshape(1, -1)
                            except Exception: # Bỏ qua embedding này nếu không reshape được
                                continue


                            for db_id, known_id_embeddings_list in facenet_db.items():
                                if not known_id_embeddings_list: continue
                                try:
                                    known_id_embeddings_array = np.array([np.asarray(e_known).flatten() for e_known in known_id_embeddings_list])
                                    if known_id_embeddings_array.ndim == 1:
                                        known_id_embeddings_array = known_id_embeddings_array.reshape(1, -1)

                                    if known_id_embeddings_array.ndim != 2 or \
                                       known_id_embeddings_array.shape[0] == 0 or \
                                       known_id_embeddings_array.shape[1] != current_emb_reshaped.shape[1]:
                                        continue
                                    sim_scores = cosine_similarity(current_emb_reshaped, known_id_embeddings_array)
                                    current_person_max_sim = np.max(sim_scores) if sim_scores.size > 0 else 0.0
                                    if current_person_max_sim > max_similarity_score:
                                        max_similarity_score = current_person_max_sim
                                        if max_similarity_score >= threshold:
                                            best_match_identity = db_id
                                except Exception: # Lỗi khi tính similarity
                                    continue

                            display_text = f"{best_match_identity} ({max_similarity_score:.2f})"
                            box_color_bgr = (0, 0, 255) # Đỏ
                            if best_match_identity != "Unknown": box_color_bgr = (0, 255, 0) # Xanh
                            cv2.rectangle(bbox_overlay_img_rgba, (fx, fy), (fx+fw, fy+fh), (*box_color_bgr, 255), 2)
                            cv2.putText(bbox_overlay_img_rgba, display_text, (fx, fy - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.45, (*box_color_bgr, 255), 2)
                            frame_recognition_labels.append(f"{best_match_identity}({max_similarity_score:.2f})")
                    # --- KẾT THÚC ĐOẠN CODE ĐÃ SỬA ---
                current_label_html = " | ".join(frame_recognition_labels) if frame_recognition_labels else \
                                     ("Không nhận dạng được." if detected_faces_info else "Đang tìm (MTCNN)...")

            alpha_channel = bbox_overlay_img_rgba[:,:,3]
            drawn_pixels_mask = np.any(bbox_overlay_img_rgba[:,:,:3] != [0,0,0], axis=2)
            alpha_channel[drawn_pixels_mask] = 255
            bbox_overlay_img_rgba[:,:,3] = alpha_channel
            current_bbox_data_url = bbox_to_bytes_cv(bbox_overlay_img_rgba)
            time.sleep(0.03)


    except KeyboardInterrupt:
        print("Nhận dạng trực tiếp bị dừng bởi người dùng (KeyboardInterrupt).")
    except Exception as e_main_loop:
        import traceback
        print(f"Lỗi nghiêm trọng trong vòng lặp nhận dạng: {type(e_main_loop).__name__}: {e_main_loop}")
        print(traceback.format_exc())
    finally:
        print("Đang cố gắng đóng luồng webcam an toàn...")
        try:
            # Gọi hàm userShutdown trong JS, hàm này sẽ tự gọi removeDom
            eval_js('userShutdown();')
            print("Đã gửi lệnh userShutdown() tới JavaScript để đóng webcam.")
        except Exception:
            print("Nếu webcam không tắt, bạn có thể cần làm mới (refresh) trang Colab.")
    print("Chương trình nhận dạng trực tiếp đã kết thúc.")


# --- PHẦN 5: THỰC THI ---
# Để chạy trong Colab:
# 1. Chạy các cell ở PHẦN 1 để khởi tạo MTCNN, FaceNet model và database.
# 2. Sau đó chạy cell này (bỏ comment phần gọi hàm):


if __name__ == "__main__":
     print("Để chạy trong Colab, hãy bỏ comment phần gọi hàm recognize_faces_live_facenet bên dưới và chạy cell đó.")
     print("Đảm bảo PHẦN 1 đã được chạy và các biến cần thiết đã được khởi tạo.")


# # === BỎ COMMENT ĐỂ CHẠY TRONG COLAB ===
# Đảm bảo các biến này được khởi tạo đúng ở Phần 1
if detector_mtcnn is not None and \
   facenet_model is not None and \
   face_database_facenet: # Kiểm tra database có dữ liệu
    print("\nBắt đầu quá trình nhận dạng khuôn mặt trực tiếp từ Colab cell...")
    recognize_faces_live_facenet(
        facenet_db=face_database_facenet,
        threshold=0.50, # Ngưỡng cosine similarity
        mtcnn_detector_obj=detector_mtcnn,
        facenet_embedding_model_obj=facenet_model
    )
else:
    print("\nKhông thể bắt đầu nhận dạng. Vui lòng kiểm tra các lỗi/cảnh báo ở PHẦN 1:")
    if detector_mtcnn is None: print("- MTCNN detector chưa được khởi tạo.")
    if facenet_model is None: print("- Mô hình FaceNet (`facenet_model`) chưa được tải/khởi tạo.")
    if not face_database_facenet: print("- Cơ sở dữ liệu khuôn mặt (`face_database_facenet`) trống hoặc chưa được tạo.")
# # === KẾT THÚC PHẦN BỎ COMMENT ===

```
